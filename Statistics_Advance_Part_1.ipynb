#1-What is a random variable in probability theory?
- In probability theory, a random variable is a mathematical object that represents a numerical outcome of a random process or experiment. It’s a way to assign numbers to the possible outcomes of an event, so we can work with them quantitatively.
Formally, a random variable is a function that maps outcomes from a sample space (the set of all possible outcomes of an experiment) to real numbers. For example, if you flip a coin, the sample space might be {heads, tails}. A random variable could assign "heads" to 1 and "tails" to 0, letting you analyze the result numerically.



#2-What are the types of random variable?
-Discrete Random Variables  
These take on a finite or countably infinite set of distinct values.  

Think of them as variables where you can list out all possible outcomes, even if the list is infinite but countable (like 1, 2, 3, ...).  

Examples:  
The number of heads in 10 coin flips (0, 1, 2, ..., 10).  

Continuous Random Variables  
These can take any value within a continuous range, like an interval of real numbers.  

The possible outcomes are uncountably infinite—you can’t list them all because they include every number in some range (e.g., 0 to 1, or all positive reals).  

Examples:  
The exact time it takes for a light bulb to burn out (e.g., 100.53 hours, 100.54 hours, etc.)



#3- What is the difference between discrete and continuous distributions?
- Discrete Distribution:  
Applies to discrete random variables, which take on a finite or countably infinite set of distinct values (e.g., 0, 1, 2, ..., or {1, 3, 5}).  

The outcomes are separate and specific—you can list them, even if the list is infinite but countable.  

Example: Number of students in a class (10, 11, 12, etc.).

Continuous Distribution:  
Applies to continuous random variables, which can take any value within a range of real numbers (e.g., any value between 0 and 1, or 5 to 10).  

The outcomes are uncountably infinite—there’s an infinite number of possibilities between any two points.  

Example: Time to failure of a machine (e.g., 3.245 hours, 3.246 hours, etc.).

How Probabilities Are Assigned
Discrete Distribution:  
Uses a probability mass function (PMF).  

The PMF gives the probability of each specific value occurring (e.g., P(X = 3) = 0.2).  

Probabilities are assigned to individual points, and the sum of all probabilities across all possible values equals 1:
∑P(X=x)=1\sum P(X = x) = 1\sum P(X = x) = 1
.  

Example: For a die roll, P(X = 1) = 1/6, P(X = 2) = 1/6, etc.

Continuous Distribution:  
Uses a probability density function (PDF).  

The PDF doesn’t give the probability of an exact value (since P(X = exact value) = 0 for continuous variables). Instead, it describes the density of probability over a range, and you calculate probabilities by integrating the PDF over an interval (e.g., P(a ≤ X ≤ b) = ∫ from a to b of f(x) dx).  

The total area under the PDF curve equals 1





#4-What are probability distribution functions (PDF)?
A probability density function (PDF) is a mathematical function used in probability theory to describe the likelihood of a continuous random variable taking on values within a certain range. It’s a core concept for continuous distributions, distinct from how discrete random variables are handled

The PDF, often denoted ( f(x) ), doesn’t give the probability of a specific value directly (unlike a probability mass function for discrete variables). Instead, it represents the density of probability at each point along the variable’s range.

To find the actual probability that the random variable ( X ) falls within an interval, say between ( a ) and ( b ), you integrate the PDF over that interval:
P(a≤X≤b)=∫abf(x) dxP(a \leq X \leq b) = \int_{a}^{b} f(x) \, dxP(a \leq X \leq b) = \int_{a}^{b} f(x) \, dx
.
This integral calculates the area under the PDF curve between ( a ) and ( b ), which is the probability.

#13- What is the Central Limit Theorem (CLT), and why is it critical in statistics?
The Central Limit Theorem (CLT) is a foundational principle in probability and statistics that explains why the normal distribution appears so often in the real world. It states that the distribution of the sum (or average) of a large number of independent, identically distributed (i.i.d.) random variables, regardless of their original distribution, tends to approximate a normal distribution as the sample size increases, provided the variables have a finite mean and variance. This convergence to normality makes the CLT a critical tool for statistical inference.



#14- How does the Central Limit Theorem relate to the normal distribution?
- The Central Limit Theorem (CLT) and the normal distribution are deeply intertwined, as the CLT explains why the normal distribution is so prevalent in statistics and provides a mechanism for approximating many real-world phenomena with it. Specifically, the CLT states that the distribution of the sum or average of a large number of independent, identically distributed (i.i.d.) random variables, regardless of their original distribution, converges to a normal distribution as the sample size increases, assuming finite mean and variance.

#18- What is the significance of confidence intervals in statistical analysis?
- Confidence intervals (CIs) are a key tool in statistical analysis, providing a range of values that likely contains an unknown population parameter (e.g., mean, proportion) along with a confidence level that quantifies the reliability of that range. Their significance lies in how they balance precision, uncertainty, and inference, making them essential for interpreting data and drawing conclusions.

Quantifying Uncertainty
What They Do: Unlike a point estimate (e.g., sample mean xˉ\bar{x}\bar{x}
), which gives a single value, a CI offers a range (e.g., 167 to 173 cm) that reflects the variability inherent in sampling. This range acknowledges that the true parameter (e.g., population mean μ\mu\mu
) isn’t exactly known but is likely within the interval.

Significance: They prevent overconfidence in a single number by showing the plausible spread of the parameter, based on sample data.

Confidence Level Interpretation
Definition: The confidence level (e.g., 95%) represents the long-run probability that the method produces an interval containing the true parameter value across repeated samples. It’s not the probability that the specific interval contains the parameter (a common misconception), but the reliability of the process.

Significance: A 95% CI means that if you took many samples and built a CI from each, about 95% of those intervals would capture the true μ\mu\mu
. This gives a standardized way to assess how trustworthy the estimate is.

#2- Generate a discrete uniform distribution using Python and plot the probability mass function (PMF)?


```python
import random
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

def generate_discrete_uniform(n, a, b):
    """Generate n samples from a discrete uniform distribution U(a, b)"""
    return [random.randint(a, b) for _ in range(n)]

def plot_pmf(samples, a, b):
    """Plot the probability mass function (PMF)"""
    counts = Counter(samples)
    x_values = np.arange(a, b + 1)
    pmf_values = [counts[x] / len(samples) for x in x_values]

    plt.bar(x_values, pmf_values, width=0.6, color='blue', alpha=0.7, edgecolor='black')
    plt.xlabel("Value")
    plt.ylabel("Probability")
    plt.title("Probability Mass Function (PMF) of Discrete Uniform Distribution")
    plt.xticks(x_values)
    plt.show()
```

#4- Write a Python script to simulate a binomial distribution with n=10 and p=0.5, then plot its histogram?

```python
import random
import matplotlib.pyplot as plt
import numpy as np
from collections import Counter

def bernoulli_pdf(x, p):
    """Calculate the probability distribution function (PDF) of a Bernoulli distribution."""
    return p if x == 1 else (1 - p)

def simulate_binomial(n, p, size):
    """Simulate a Binomial distribution."""
    return np.random.binomial(n, p, size)

def plot_binomial_histogram(samples, n, p):
    """Plot histogram of a Binomial distribution."""
    plt.hist(samples, bins=n+1, range=(0, n+1), density=True, alpha=0.7, color='blue', edgecolor='black')
    plt.xlabel("Number of Successes")
    plt.ylabel("Probability")
    plt.title(f"Binomial Distribution (n={n}, p={p})")
    plt.show()
n, p, size = 10, 0.5, 1000  # n trials, probability p, sample size

# Simulate and plot
binomial_samples = simulate_binomial(n, p, size)
plot_binomial_histogram(binomial_samples, n, p)

# Example usage of Bernoulli PDF
p = 0.5  # Probability of success
print(f"P(X=1): {bernoulli_pdf(1, p)}")
print(f"P(X=0): {bernoulli_pdf(0, p)}")
```

#6- Write a Python program to calculate and plot the cumulative distribution function (CDF) of a discrete uniform distribution?

```python
a = 1  # lower bound
b = 6  # upper bound
n = b - a + 1  # number of possible values

# Create x values (all possible outcomes)
x = np.arange(a, b + 1)

# Calculate CDF: probability accumulates as 1/n for each step
cdf = np.arange(1, n + 1) / n

# Plot
plt.step(x, cdf, where='post')
plt.title('CDF of Discrete Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()
```

#9- Write a Python function to calculate Z-scores from a dataset and plot them.

```python
import numpy as np
import matplotlib.pyplot as plt

def calculate_and_plot_zscores(data):
    # Calculate mean and standard deviation
    mean = np.mean(data)
    std = np.std(data)

    # Calculate Z-scores
    z_scores = (data - mean) / std

    # Plot
    plt.hist(z_scores, bins=20, edgecolor='black', density=True)
    plt.title('Z-scores Distribution')
    plt.xlabel('Z-score')
    plt.ylabel('Density')
    plt.show()

    return z_scores
data = np.random.normal(10, 2, 1000)  # Sample data
z_scores = calculate_and_plot_zscores(data)
```

#10-Implement the Central Limit Theorem (CLT) using Python for a non-normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
sample_size = 50  # size of each sample
num_samples = 1000  # number of samples

population = np.random.uniform(0, 1, 10000)
sample_means = []


for _ in range(num_samples):
    sample = np.random.choice(population, sample_size)
    sample_means.append(np.mean(sample))


plt.hist(sample_means, bins=30, edgecolor='black', density=True)
plt.title('Distribution of Sample Means (CLT)')
plt.xlabel('Sample Mean')
plt.ylabel('Density')
plt.show()
```

#12- Write a Python function to calculate and plot the standard normal distribution (mean = 0, std = 1)

```python
import matplotlib.pyplot as plt

def plot_standard_normal():
    # Generate x values
    x = np.linspace(-4, 4, 100)

    # Calculate standard normal PDF: (1/√(2π)) * e^(-x²/2)
    pdf = (1 / np.sqrt(2 * np.pi)) * np.exp(-0.5 * x**2)

    # Plot
    plt.plot(x, pdf)
    plt.title('Standard Normal Distribution (μ=0, σ=1)')
    plt.xlabel('x')
    plt.ylabel('Probability Density')
    plt.grid(True)
    plt.show()
plot_standard_normal()
```

#13- Generate random variables and calculate their corresponding probabilities using the binomial distribution

```python
import numpy as np
from scipy.stats import binom
import matplotlib.pyplot as plt

# Parameters
n = 10  # number of trials
p = 0.3  # probability of success
size = 1000  # number of random variables

# Generate random variables
data = np.random.binomial(n, p, size)

# Calculate probabilities for k = 0 to n
k_values = np.arange(0, n + 1)
probabilities = binom.pmf(k_values, n, p)

# Plot histogram of generated data
plt.hist(data, bins=range(n + 2), edgecolor='black', density=True, alpha=0.7, label='Simulated')
# Plot theoretical probabilities
plt.plot(k_values, probabilities, 'ro-', label='Theoretical')
plt.title(f'Binomial Distribution (n={n}, p={p})')
plt.xlabel('Number of Successes')
plt.ylabel('Probability')
plt.legend()
plt.show()
```

#14-Write a Python program to calculate the Z-score for a given data point and compare it to a standard normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Sample data and data point
data = np.random.normal(50, 10, 1000)  # mean=50, std=10
x = 60  # data point to calculate Z-score for

# Calculate Z-score
mean = np.mean(data)
std = np.std(data)
z_score = (x - mean) / std

# Generate standard normal distribution for comparison
x_values = np.linspace(-4, 4, 100)
std_normal = norm.pdf(x_values, 0, 1)

# Plot
plt.plot(x_values, std_normal, label='Standard Normal')
plt.axvline(z_score, color='red', linestyle='--', label=f'Z-score = {z_score:.2f}')
plt.title('Z-score vs Standard Normal Distribution')
plt.xlabel('Z-value')
plt.ylabel('Probability Density')
plt.legend()
plt.grid(True)
plt.show()

print(f"Data point: {x}")
print(f"Z-score: {z_score:.2f}")

```

#15- Implement hypothesis testing using Z-statistics for a sample dataset

```python
import numpy as np
from scipy.stats import norm

# Sample data and parameters
sample = np.random.normal(100, 15, 50)  # sample data
pop_mean = 95  # null hypothesis population mean
alpha = 0.05  # significance level

# Calculate sample statistics
sample_mean = np.mean(sample)
sample_std = np.std(sample, ddof=1)  # sample standard deviation
n = len(sample)

# Calculate Z-statistic
z_stat = (sample_mean - pop_mean) / (sample_std / np.sqrt(n))

# Calculate p-value (two-tailed test)
p_value = 2 * (1 - norm.cdf(abs(z_stat)))

# Critical value for two-tailed test
z_critical = norm.ppf(1 - alpha/2)

# Results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"Z-statistic: {z_stat:.2f}")
print(f"P-value: {p_value:.4f}")
print(f"Critical Z-value: ±{z_critical:.2f}")

# Decision
if abs(z_stat) > z_critical or p_value < alpha:
    print("Reject the null hypothesis")
else:
    print("Fail to reject the null hypothesis")
```

#16- Create a confidence interval for a dataset using Python and interpret the result

```python
import numpy as np
from scipy.stats import t

# Sample data
data = np.random.normal(50, 10, 100)  # mean≈50, std≈10, n=100
confidence_level = 0.95  # 95% confidence

# Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)  # sample standard deviation
n = len(data)

# Calculate margin of error (using t-distribution)
t_critical = t.ppf((1 + confidence_level) / 2, df=n-1)
margin_error = t_critical * (sample_std / np.sqrt(n))

# Calculate confidence interval
ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

# Results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print(f"Margin of Error: ±{margin_error:.2f}")

# Interpretation
print("\nInterpretation:")
print(f"We are 95% confident that the true population mean lies between {ci_lower:.2f} and {ci_upper:.2f}.")
```

#17- Generate data from a normal distribution, then calculate and interpret the confidence interval for its mean

```python
import numpy as np
from scipy.stats import t

# Generate data from normal distribution
mu = 75  # true mean
sigma = 12  # true standard deviation
n = 50  # sample size
data = np.random.normal(mu, sigma, n)

# Calculate sample statistics
sample_mean = np.mean(data)
sample_std = np.std(data, ddof=1)
confidence_level = 0.95

# Calculate confidence interval
t_critical = t.ppf((1 + confidence_level) / 2, df=n-1)
margin_error = t_critical * (sample_std / np.sqrt(n))
ci_lower = sample_mean - margin_error
ci_upper = sample_mean + margin_error

# Results
print(f"Sample Mean: {sample_mean:.2f}")
print(f"95% Confidence Interval: ({ci_lower:.2f}, {ci_upper:.2f})")
print(f"Margin of Error: ±{margin_error:.2f}")

# Interpretation
print("\nInterpretation:")
print(f"We are 95% confident that the true mean of the population is between {ci_lower:.2f} and {ci_upper:.2f}.")
print(f"The true mean ({mu}) is {'within' if ci_lower <= mu <= ci_upper else 'outside'} this interval.")
```

#19- Write a Python script to calculate and visualize the probability density function (PDF) of a normal distribution

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 0  # mean
sigma = 1  # standard deviation

# Generate x values
x = np.linspace(mu - 4*sigma, mu + 4*sigma, 100)

# Calculate PDF: (1/(σ√(2π))) * e^(-(x-μ)²/(2σ²))
pdf = (1 / (sigma * np.sqrt(2 * np.pi))) * np.exp(-((x - mu)**2) / (2 * sigma**2))

# Plot
plt.plot(x, pdf)
plt.title(f'Normal Distribution PDF (μ={mu}, σ={sigma})')
plt.xlabel('x')
plt.ylabel('Probability Density')
plt.grid(True)
plt.show()
```

#20- Use Python to calculate and interpret the cumulative distribution function (CDF) of a Poisson distribution

```python
import numpy as np
from scipy.stats import poisson
import matplotlib.pyplot as plt

# Parameter
lambda_val = 3  # average rate

# Generate x values (number of events)
x = np.arange(0, 11)

# Calculate CDF
cdf = poisson.cdf(x, lambda_val)

# Plot
plt.step(x, cdf, where='post')
plt.title(f'Poisson CDF (λ={lambda_val})')
plt.xlabel('Number of Events')
plt.ylabel('Cumulative Probability')
plt.grid(True)
plt.show()
k = 4
prob = poisson.cdf(k, lambda_val)
print(f"Probability of {k} or fewer events: {prob:.4f}")
```

#22- Write a Python program to compare the standard deviations of two datasets and visualize the difference

```python
import numpy as np
import matplotlib.pyplot as plt

def compare_std(data1, data2):
    std1 = np.std(data1)
    std2 = np.std(data2)
    print(f"Standard Deviation of Dataset 1: {std1:.2f}")
    print(f"Standard Deviation of Dataset 2: {std2:.2f}")

    plt.boxplot([data1, data2], labels=['Dataset 1', 'Dataset 2'])
    plt.title("Comparison of Standard Deviations")
    plt.ylabel("Values")
    plt.show()

# Example datasets
data1 = np.random.normal(50, 5, 100)
data2 = np.random.normal(50, 15, 100)

compare_std(data1, data2)
```

#23- Calculate the range and interquartile range (IQR) of a dataset generated from a normal distribution

```python
import numpy as np

# Generate dataset from a normal distribution
data = np.random.normal(loc=50, scale=10, size=1000)

# Calculate range
data_range = np.max(data) - np.min(data)

# Calculate IQR
q1 = np.percentile(data, 25)
q3 = np.percentile(data, 75)
iqr = q3 - q1

# Output
print(f"Range: {data_range:.2f}")
print(f"IQR: {iqr:.2f}")
```

#24-Implement Z-score normalization on a dataset and visualize its transformation

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate a dataset
data = np.random.normal(loc=50, scale=10, size=1000)

# Z-score normalization
mean = np.mean(data)
std = np.std(data)
z_scores = (data - mean) / std

# Visualization
plt.figure(figsize=(10, 4))

# Original data
plt.subplot(1, 2, 1)
plt.hist(data, bins=30, color='skyblue', edgecolor='black')
plt.title("Original Data")
plt.xlabel("Value")
plt.ylabel("Frequency")

# Normalized data
plt.subplot(1, 2, 2)
plt.hist(z_scores, bins=30, color='lightgreen', edgecolor='black')
plt.title("Z-score Normalized Data")
plt.xlabel("Z-score")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```

#25- Write a Python function to calculate the skewness and kurtosis of a dataset generated from a normal distribution.

```python
import numpy as np
from scipy.stats import skew, kurtosis

def calculate_skewness_kurtosis(data):
    skewness = skew(data)
    kurt = kurtosis(data)  # default is Fisher's definition (normal ==> 0)
    print(f"Skewness: {skewness:.2f}")
    print(f"Kurtosis: {kurt:.2f}")

# Generate dataset
data = np.random.normal(loc=0, scale=1, size=1000)

# Call the function
calculate_skewness_kurtosis(data)
```
